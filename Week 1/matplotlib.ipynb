```python
import numpy as np
import matplotlib.pyplot as plt
m = np.linspace(0,10,11)
m
```

```python
c = 3 * (10 ** 8)

E = m * c**2
E
```

```python
c = 3e8  # speed of light

plt.plot(m, E, color='red', linewidth=3)
plt.title('E=mc^2')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
plt.xlim(0, 10)
plt.ylim(0, 9e17)
plt.show()
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 = [4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
plt.plot(labels, july16_2007, label='july16_2007')
plt.plot(labels, july16_2020, label='july16_2020')
plt.legend()
plt.show()
```

```python
plt.plot(labels, july16_2007, label='july16_2007')
plt.plot(labels, july16_2020, label='july16_2020')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')   # legend ko box ke bahar nikal diya
plt.subplots_adjust(right=0.8)                            # figure mein legend ke liye space banaya
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

axes[0].plot(labels, july16_2007)
axes[0].set_title('July 16th, 2007')

axes[1].plot(labels, july16_2020)
axes[1].set_title('July 16th, 2020')

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(labels, july16_2007, color='blue')
ax1.set_ylabel('2007', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, color='red')
ax2.set_ylabel('2020', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('July 16th Yield Curves')
plt.show()
```
