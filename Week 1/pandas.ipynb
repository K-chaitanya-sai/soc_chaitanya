```python
import pandas as pd
```

```python
hotels = pd.read_csv(r"C:\Users\lucka\Downloads\hotel_booking_data.csv")
```

```python
hotels.head()
```

```python
# 1. How many rows are there?
len(hotels)
```

```python
# 2. Missing data and column with most missing values
hotels.isnull().sum().sort_values(ascending=False)
```

```python
# 3. Drop the company column
hotels = hotels.drop("company", axis=1)
```

```python
# 4. Top 5 most common country codes
hotels["country"].value_counts().head()
```

```python
# 5. Person who paid the highest ADR
hotels.loc[hotels["adr"].idxmax(), ["name", "adr"]]
```

```python
# 6. Mean ADR
hotels["adr"].mean()
```

```python
# 7. Average number of nights
(
    hotels["stays_in_week_nights"] +
    hotels["stays_in_weekend_nights"]
).mean().round(2)
```

```python
# 8. Average total cost of stay
(
    hotels["adr"] *
    (hotels["stays_in_week_nights"] + hotels["stays_in_weekend_nights"])
).mean().round(2)
```

```python
# 9. Names and emails with exactly 5 special requests
hotels.loc[
    hotels["total_of_special_requests"] == 5,
    ["name", "email"]
]
```

```python
# 10. Percentage of repeat guests
(hotels["is_repeated_guest"].mean() * 100).round(2)
```

```python
# 11. Top 5 most common last names
hotels["name"].str.split().str[-1].value_counts().head()
```

```python
# 12. People who booked the most children and babies
family = hotels["children"] + hotels["babies"]

hotels.loc[
    family == family.max(),
    ["name", "children", "babies"]
]
```

```python
# 13. Top 3 most common phone area codes
hotels["phone-number"].astype(str).str[:3].value_counts().head(3)
```

```python
# 14. Arrivals between 1st and 15th (inclusive)
hotels[
    (hotels["arrival_date_day_of_month"] >= 1) &
    (hotels["arrival_date_day_of_month"] <= 15)
].shape[0]
```

```python
# 15. Count arrivals by day of the week (Hard Bonus)
import pandas as pd

hotels["arrival_day"] = pd.to_datetime(
    hotels["arrival_date_year"].astype(str) + "-" +
    hotels["arrival_date_month"] + "-" +
    hotels["arrival_date_day_of_month"].astype(str)
)

hotels["arrival_day"].dt.day_name().value_counts()
```
